```python
%matplotlib inline
import pandas as pd
df = pd.read_csv("population.csv", encoding="SHIFT_JIS")
df["増減"] = df["平成28年"]-df["平成22年"]
df["test"] = 0
df = df.sort_values(by="増減",ascending=False)
top10 = df[0:10]
top10.plot.bar(y="増減",x="都道府県")
top10
```
